```python
# Adjascency List representation in Python
class Graph:
    V = 0   #No of Vertices
    graphDict = {}     #dicionary to store adjascency list

    #Constructor
    def __init__(self, _graphDict = None):
        if _graphDict is None:
          _graphDict = {}
        self.V = len(_graphDict)
        self.graphDict = _graphDict
        
    def takeInput(self, n, w):
        if n in self.graphDict.keys():
          self.graphDict[n].append(w)
          if w not in self.graphDict.keys():
            self.graphDict[w] = []
            self.V += 1
        else:
          self.graphDict[n] = [w]
          self.V += 1
          if w not in self.graphDict.keys():
            self.graphDict[w] = []
            self.V += 1


    def getVertices(self):
        return list(self.graphDict.keys())
    # Print the graph
    def printGraph(self):
        print("Graph G:")
        for key in self.graphDict.keys():
            print(str(key) + "  -> " + str(self.graphDict[key]))
    
    def readGraph(self,path):
      file = open(path,'r')
      text = file.read()
      text = text.replace(" ", "")
      lines = text.split('\n')
      lines.pop(0)
      for i in range(len(lines)):
        splitted = lines[i].split(':')
        vertices = splitted[1].split(',')
        vertex = int(splitted[0])
        index = 0
        for j in vertices:
          if int(j) == 1:
            self.takeInput(vertex,index)
          index += 1
      file.close()

    #print the paths from start to end vertex generated by the DFS method.
    def exploreFunction(self,start, end):
      #visited
      visited = [False]*self.V

      #paths
      path = []
      paths = []

      #utility Function
      self.exploreFunctionUtility(start,end,visited,path,paths)

      #print paths
      print("Paths from start to end vertex:\n " + str(paths))

    def exploreFunctionUtility(self,start,end,visited,path,paths):
      visited[start] = True
      path.append(start)

      if start == end:
        paths.append(list(path))
      else:
        for vertex in self.graphDict[start]:
          if visited[vertex] == False:
            self.exploreFunctionUtility(vertex,end,visited,path,paths)

      path.pop()
      visited[start] = False

def main():
  g = Graph()
  g.readGraph("/content/drive/MyDrive/DataSets/AdjacencyList.txt")
  # g.takeInput(0, 1)
  # g.takeInput(0, 2)
  # g.takeInput(1, 2)
  # g.takeInput(2, 0)
  # g.takeInput(2, 3)
  # g.takeInput(3, 3)
  g.printGraph()
  g.exploreFunction(2,3)

if __name__ == "__main__":
  main()
```
